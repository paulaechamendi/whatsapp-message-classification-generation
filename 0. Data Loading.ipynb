```python
import pandas as pd
import re

# Cargar el archivo de conversación
with open('/content/drive/MyDrive/TRABAJO_NO_ESTRUCTURADOS/data/chat_grupo_raw.txt', 'r', encoding='utf-8') as file:
    chat_data = file.readlines()

# Nueva expresión regular para extraer la fecha, persona y mensaje
pattern = r'\[(\d{1,2}/\d{1,2}/\d{2}), (\d{2}:\d{2}:\d{2})\]\s([A-Za-z\s]+):\s(.+)'

# Crear una lista con las tuplas (fecha, persona, mensaje)
messages = []
for line in chat_data:
    match = re.match(pattern, line)
    if match:
        date, time, person, message = match.groups()
        # Unir la fecha y la hora para obtener el timestamp completo
        timestamp = f"{date} {time}"
        messages.append((timestamp, person, message))

# Crear un DataFrame con las tres columnas
df = pd.DataFrame(messages, columns=['timestamp', 'person', 'message'])

# Verificar los primeros registros
df.head(20)

# Guardar el DataFrame limpio en un archivo CSV
df.to_csv('/content/drive/MyDrive/TRABAJO_NO_ESTRUCTURADOS/data/dataframe_chat.csv', index=False)
```
